```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.1;
NS = 10;

cfl = 0.5;

ta = cfl*h*h;
N = round(1/ta)
```

```python
xxx = np.linspace(h,h*NS,NS)

a = np.zeros((NS, NS))
np.fill_diagonal(a,-2*cfl)

u = np.zeros((NS, N))

for k in range(0,NS):
    u[k,0] = np.sin((k+1)*h*np.pi)
    
    
    
```

```python
vvv=[]
ttt=[]

u0 = u[:,0]
vvv.append(max(u0))
ttt.append(0)

for k in range(0,NS-1):
    a[k,k+1] = cfl;
    a[k+1,k] = cfl;
```

```python
b = np.eye(NS) - cfl*a

for kk in range(1,N):
    u1 = np.linalg.solve(b,u0)
    u[:,kk] = u1
    u0 = u1
    vvv.append(max(u0))
    ttt.append( ttt[kk-1] + ta)

    
    
```

```python
fig = plt.figure()
plt.plot(ttt, vvv, '-')
plt.xlabel('time t')
plt.ylabel('max option price u')
```

```python
from mpl_toolkits.mplot3d import Axes3D  
# Axes3D import has side effects, it enables using projection='3d' in add_subplot


fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')

X, Y = ttt, xxx
Z = u

ax.plot_surface(X, Y, Z)

ax.set_xlabel('time t')
ax.set_ylabel('value s')
ax.set_zlabel('option price u')

plt.show()
```
